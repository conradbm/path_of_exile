```python
# import libraries
from __future__ import division
import urllib2
from bs4 import BeautifulSoup
import sys
import os
import subprocess
import re
import threading
import time
import datetime
import MySQLdb

hostname = '127.0.0.1'
username = 'root'
password = 'Jesusislife1!'
database = 'PathOfExileDB'
       

def run(interval, url, conn):
    
    # Scrape the url -- raw html returned
    soup = scrape_page(url)
    
    # Convert the BeautifulSoup object to a string
    target_string = str(soup)

    # We need the following fields:
        # $1 buy_currency
        # $2 buy_value
        # $4 sell_currency
        # $5 sell_value
        
    # Regex to get the contents from the div we see repeated with the values we want
    result_set = re.finditer(r'<div class=\"displayoffer \" data-buycurrency=\"(.*?)\" data-buyvalue=\"(.*?)\" data-ign=\"(.*?)\" data-sellcurrency=\"(.*?)\" data-sellvalue=\"(.*?)\" (?:data-stock=\"(\d*)\")* data-username=\"(.*?)\">', target_string)                     
    results = map(lambda x: x.groups(),result_set)

    for res in results:
        
        # Print contents to confirm
        print res[0], "\t", res[1], "\t", res[3], "\t", res[4]
        
        # Make a cursor
        cur = conn.cursor()
        
        # Construct SP statement
        insertQuery = "CALL SP_INSERT_INTO_CURRENCY(" +res[0] + ","+res[1] +","+res[3]+","+res[4]+");"
        
        # Execute our SP
        cur.execute(insertQuery)   
    
    # Wait until the specified interval is over
    time.sleep(interval) # SECONDS!!!
    
def scrape_page(some_url):
    quote_page = some_url
    page = urllib2.urlopen(quote_page)
    soup = BeautifulSoup(page, "html.parser")
    return soup

def main():
    ### Run this: python scrape.py <interval> <target>
    interval = 60 * 10
    site = "http://currency.poe.trade/search?league=Legacy&online=x&want=&have=5"
    print "Interval (mins): ", str(interval)
    print "Target URL: ", site
    
    # Create a static DB connection once
    
    myConnection = MySQLdb.connect( host=hostname, user=username, passwd=password, db=database )
    while True:
        run(interval, site, myConnection)
    myConnection.close()

main()
```
